### Section 1: Background & Problem Statement

#### Section 1.1: Background

Peer-to-peer (P2P) lending enables individuals to obtain loans directly from other individuals, cutting out the financial institution as the middleman and this sometimes known as "social lending". One of such companies is LendingClub in United States where connect borrowers directly to investors. Borrowers would fill out an application detailing their credit history, loan details, employment status and other self-reported information by which Lending Club would assign a loan grade reflecting the quality of a loan. Potential investors would be able to view the details of various loan applications and make a final decision to invest in the loan.

#### Section 1.2: Problem Statement

The main object of P2P lenders are individual investors who want to get the best risk-reward return for their investment. Our aim for this project is to help P2P lenders decide on the which loans to invest in their portfolio to obtain the best risk-reward (Sharpe Ratio) returns.

```python
# function to evaluate model
def evaluate_model(model_name,model,X_train,y_train,X_test,y_test):
    train_auc = np.mean(cross_val_score(model, X_train, y_train, scoring = 'roc_auc', cv = 5))
    preds = model.predict(X_test)
    test_auc = roc_auc_score(y_test, preds)
    precision, recall, fscore, support = precision_recall_fscore_support(y_test, preds,average='weighted')
    train_accuracy = model.score(X_train, y_train)
    test_accuracy = model.score(X_test,y_test)
    summary1 = {'Model_name' : model_name,
                'train_accuracy': round(train_accuracy,3),
                'test_accuracy': round(test_accuracy,3),
                'train_auc': round(train_auc,3),
                'test_auc': round(test_auc,3),
                'Precision': round(precision,3),
                'recall': round(recall,3),
                'fscore' : round(fscore,3)}
    summary= pd.DataFrame([summary1])
    return summary
```

```python
# Instantiate a NN.
nn = Sequential()

# Input layer
nn.add(Dense(X_under.shape[1], activation='relu', input_dim=X_under.shape[1]))
nn.add(Dropout(0.2))

nn.add(Dense(40, activation='relu'))
nn.add(Dropout(0.2))

nn.add(Dense(24, activation='relu'))
nn.add(Dropout(0.2))
nn.add(Dense(16, activation='relu'))
nn.add(Dropout(0.2))

# Output layer
nn.add(Dense(1, activation='sigmoid')) 

# Compile it
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['AUC'])

early_stop = EarlyStopping(monitor='val_auc', min_delta=0, patience=2, mode='auto')

nn.fit(X_under, y_under, batch_size=256, validation_data=(X_scaled_test, y_test),
                          epochs=25,
                          callbacks=[early_stop],
                          verbose=1)
```
